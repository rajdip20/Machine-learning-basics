<B> <p style="color:brown;font-size:15px">(ii) Make a structure for how many matches played in city(ig.Kolkata/Mumbai)?

```python
def num_matches(city):
    mask=match['city']==city
    return match[mask].shape[0]
```

<B> <p style="color:brown;font-size:15px">(iv) Make a structure for how many matches played in a particular ground (ig.Eden Gardens/Newlands) (2008-2017)?

```python
def ground(venue):
    mask=match['venue']==venue
    return match[mask].shape[0]
```

<B> <p style="color:brown;font-size:15px">(v) Make a structure for how many matches played in a city in a particular year (2008-20017)?

```python
def city_year(city,season):
    s1=match['season']==season
    s2=match['city']==city
    s3=match[s1 & s2].shape[0]
    return s3
```

```python
city_year('Mumbai',2015)
```

```python
city_year('Kolkata',2010)
```

<B> <p style="color:brown;font-size:15px">(xix)Make a structure display the team's name which are give more run to <I>Input</I> batsman?

```python
def batsman_v_team(batsman_name):
    mask=delivery[delivery['batsman']==batsman_name]
    return mask.groupby('bowling_team').sum()['batsman_runs'].sort_values(ascending=False).head(1).index[0]
```

<B> <p style="color:brown;font-size:15px">(xxi) Make a structure to display career graph between two <I>Input</I> batsman (2008-2017)?

```python
def career_graph(batsman_name1,batsman_name2):
    batsman_df1=total[total['batsman']==batsman_name1]
    batsman_df1.groupby('season').sum()['batsman_runs'].plot()
    batsman_df2=total[total['batsman']==batsman_name2]
    batsman_df2.groupby('season').sum()['batsman_runs'].plot()
```

<B> <p style="color:brown;font-size:15px">(xxxvii) Make a new column by the name of 'family_type' and it cointain with family type(ig.small,medium,large) besis of size ?

```python
def family_label(fsize):
    if fsize==0:
        return "Alone"
    elif fsize>0 and fsize<=3:
        return "Small"
    elif fsize>3 and fsize<=6:
        return "Medium"
    else:
        return "Large"
titanic['family_type']=titanic['Family_size'].apply(family_label)
```
